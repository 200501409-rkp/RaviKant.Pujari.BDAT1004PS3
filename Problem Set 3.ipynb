### Step 1. Import the necessary libraries

````python
import numpy as np
import pandas as pd
````

### Step 2. Import the dataset from this address
### Step 3. Assign it to a variable called users

````python
# dataset is used from the given link (https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user)

# imported dataset is assigned to variable called 'users'
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|') 
````

### Step 4. Discover what is the mean age per occupation

````python
# first we need to group the dataset by 'occupation' and next find the mean

# Method 1 -> Directly referring to the column name

mean_age_per_occupation = users[['occupation', 'age']].groupby('occupation').mean()

# Method 2 -> Referring to column by position or location using 'iloc'

# mean_age_per_occupation = users.iloc[:, [3, 1]].groupby('occupation').mean()

mean_age_per_occupation # display the result -> mean age in each occupation
````

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

````python
# Let's extract gender and occupation from the dataset 
df = users[['gender', 'occupation']]

# Now, filter 'male' from gender in 'df'
males = df.loc[df['gender'] == 'M']

# Group by Occupation and count in df
group_by_occupations = df.groupby('occupation').count()

# Group by Occupation and count in males
male_count_by_occupation = males.groupby('occupation').count()

# Let's find male_ratio_per_occupation
male_ratio_per_occupation = male_count_by_occupation / group_by_occupations

male_ratio_per_occupation.sort_values(ascending=False, by='gender')
````

### Step 6. For each occupation, calculate the minimum and maximum ages

````python
df = users[['occupation', 'age']]

occupations_min_age = df.groupby('occupation').min()

occupations_max_age = df.groupby('occupation').max()

result = pd.concat([occupations_max_age, occupations_min_age], axis=1)

result
````

### Step 7. For each combination of occupation and sex, calculate the mean age

````python
df = users[['occupation', 'gender', 'age']]

users_mean_age_by_occupation_and_gender = df.groupby(['occupation', 'gender']).mean('age')

users_mean_age_by_occupation_and_gender
````

### Step 8. For each occupation present the percentage of women and men

````python
gender_percentage_by_occupation = df.groupby('occupation')['gender'].value_counts(normalize=True)*100

gender_percentage_by_occupation
````
